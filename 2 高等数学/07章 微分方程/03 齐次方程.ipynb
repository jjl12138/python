```python
from sympy import *
init_printing()
x=Symbol('x')
y=Function('y')
eq=y(x)**2+(x**2-x*y(x))*(y(x).diff(x))
dsolve(eq,y(x))
```

```python
y=symbols('y',positive=True)
x=Function('x')
eq=x(y).diff(y)-x(y)/y+sqrt(1+(x(y)/y)**2)
result=dsolve(eq,x(y))
result
```

```python
from scipy.integrate import odeint
import numpy as np 
import matplotlib.pyplot as plt
def dydx(y,x):
    return -(2*x+y-4)/(x+y-1)
x=np.linspace(0,15,100)
y=odeint(dydx,10,x) #x=0时令y=10，即本题结果中的C=80
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
