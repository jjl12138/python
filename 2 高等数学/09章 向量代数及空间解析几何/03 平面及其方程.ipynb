```python
import numpy as np 
import matplotlib.pyplot as plt
#导入三维坐标系
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-3,3,300)
y=np.linspace(-2,2,200)
xx,yy=np.meshgrid(x,y)
zz=(8-xx+2*yy)/3
fig=plt.figure()
ax=Axes3D(fig)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
#绘制zz=f(xx,yy)确定的曲面(这里为平面)
ax.plot_surface(xx,yy,zz,color='g',alpha=0.9)    # alpha用来改变图像的透明度，属于0-1
plt.show()
```

```python
#深入理解np.meshgrid(x,y)函数
x=np.linspace(1,3,3)
y=np.linspace(4,7,4)
xx,yy=np.meshgrid(x,y)
print('{}\n*************\n{}'.format(xx,yy))
```
