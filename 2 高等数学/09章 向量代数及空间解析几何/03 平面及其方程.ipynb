```python
import numpy as np 
import matplotlib.pyplot as plt
#å¯¼å…¥ä¸‰ç»´åæ ‡ç³»
from mpl_toolkits.mplot3d import Axes3D
x=np.linspace(-3,3,300)
y=np.linspace(-2,2,200)
xx,yy=np.meshgrid(x,y)
zz=(8-xx+2*yy)/3
fig=plt.figure()
ax=Axes3D(fig)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
#ç»˜åˆ¶zz=f(xx,yy)ç¡®å®šçš„æ›²é¢(è¿™é‡Œä¸ºå¹³é¢)
ax.plot_surface(xx,yy,zz,color='g',alpha=0.9)    # alphaç”¨æ¥æ”¹å˜å›¾åƒçš„é€æ˜Žåº¦ï¼Œå±žäºŽ0-1
plt.show()
```
