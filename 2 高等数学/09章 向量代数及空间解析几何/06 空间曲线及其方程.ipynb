```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
a=3.0
u=np.linspace(0,2*np.pi+0.1,100)
x=a/2*(1+np.cos(u))
y=a/2*np.sin(u)
z=np.sqrt(a**2-x**2-y**2)
fig=plt.figure()
ax=Axes3D(fig)
ax.plot(x,y,z,'r',linewidth=2.0)
plt.show()
```

```python
a,b=3.0,1.0
theta=np.linspace(0,6*np.pi,300)
x=a*np.cos(theta)
y=a*np.sin(theta)
z=b*theta
fig=plt.figure()
ax=Axes3D(fig)
ax.plot(x,y,z)
plt.show()
```

```python
from sympy.plotting import plot3d_parametric_line
from sympy import Symbol,sin,cos,pi
theta=Symbol('theta')
a,b=3,1
plot3d_parametric_line(a*sin(theta),a*cos(theta),b*theta,(theta,0,6*pi))
```
