```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
t=np.linspace(0,5)
x=1+4*t
y=-t
z=-2-3*t
fig=plt.figure()
#åˆ›å»ºä¸‰ç»´åæ ‡ç³»
ax=Axes3D(fig)
#åœ¨ä¸‰ç»´åæ ‡ä¸‹ç»˜åˆ¶æŠ˜çº¿(è¿™é‡Œæ˜¯ç›´çº¿)
ax.plot(x,y,z,'r',linewidth=3.0)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Line of (x-1)/4=y/(-1)=z/(-3)')
plt.show()
```
