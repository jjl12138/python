```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
t=np.linspace(0,5)
x=1+4*t
y=-t
z=-2-3*t
fig=plt.figure()
#创建三维坐标系
ax=Axes3D(fig)
#在三维坐标下绘制折线(这里是直线)
ax.plot(x,y,z,'r',linewidth=3.0)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Line of (x-1)/4=y/(-1)=z/(-3)')
plt.show()
```
