```python
from sympy import limit,oo,symbols,Abs
x=symbols('x')
n=symbols('n',positive=True)  #限定n值不能为负
def f(x):
    return limit((1-Abs(x)**(2*n))/(1+Abs(x)**(2*n)),n,oo)  #计算机不计算底为负数的指数函数！
epsilon=1e-5
print((f(1-epsilon),f(1),f(1+epsilon)))
print((f(-1-epsilon),f(-1),f(-1+epsilon)))
```
