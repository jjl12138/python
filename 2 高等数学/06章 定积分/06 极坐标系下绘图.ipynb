```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#例6.17
```

```python
theta=np.linspace(0,2*np.pi,100)
r=[1.0]*100
#polar()实现极坐标系下绘图
plt.polar(theta,r,linewidth=3)
plt.show()
```

```python
#例6.18
```

```python
theta=np.linspace(-np.pi/2,np.pi/2)
r=np.cos(theta)
plt.polar(theta,r,'g--')
plt.show()
```

```python
#例6.19
```

```python
theta=np.linspace(0,np.pi)
r=np.sin(theta)
plt.polar(theta,r,'r*')
plt.show()
```

```python
#例6.20
```

```python
a=2.0
theta=np.linspace(0,2*np.pi,100)
#阿基米德螺线
r=a*theta
plt.polar(theta,r,'r',linewidth=2)
plt.show()
```

```python
#例6.21
```

```python
fig=plt.figure()
#极坐标系下绘图的另一种方法
ax=fig.add_subplot(111,projection='polar')
theta=np.linspace(0,2*np.pi,200)
a=2.0
#心形线
r=a*(1+np.cos(theta))
ax.plot(theta,r,'r',linewidth=2.0,label='Cardioid')
ax.legend()
plt.show()
```
