```python
import numpy as np 
import matplotlib.pyplot as plt
def f(x):
    return np.power(np.e,-x**2/2)/((2*np.pi)**0.5)
x=np.linspace(-3,3,200)
plt.plot(x,f(x))
plt.title('An Important Curve')
plt.show()
```

```python
#生成点的个数
gen_size=100000
np.random.seed(0) #为np.random对象设置种子，以便每次运行的结果一样
x=np.random.random(size=gen_size)  #生成[0,1)之间的100000个随机数
x=6*x-3 #随机数范围变为[-3,3)
y=np.random.random(size=gen_size)/((2*np.pi)**0.5)  #[0,f(0))之间的100000个随机数
#计算gen_size个点中有多少个点在曲线下方
nums=0
for i in range(gen_size):
    if y[i]<f(x[i]):
        nums+=1.0
#矩形S的面积为6*f(0),曲线下方与x轴上方所围区域的面积估值为：S*nums/gen_size
S=6/((2*np.pi)**0.5)
S*nums/gen_size
```
