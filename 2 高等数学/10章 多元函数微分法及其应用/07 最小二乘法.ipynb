```python
import numpy as np 
import matplotlib.pyplot as plt 
import sympy as sy
```

```python
#例10.25
```

```python
t=np.array([0,1,2,3,4,5,6,7])
y=np.array([27,26.8,26.5,26.3,26.1,25.7,25.3,24.8])
#显示网格
plt.grid()
plt.scatter(t,y,c='r',s=40)
plt.show()
```

```python
a,b=sy.symbols('a b',real=True)
M=np.sum((y-(a*t+b))**2)
#求斜率及截距的值
solves=sy.solve([M.diff(a),M.diff(b)],[a,b])
solves
```

```python
k,d=np.round(float(solves[a]),3),np.round(float(solves[b]),3)
plt.grid()
plt.scatter(t,y,c='r',s=40)
plt.plot(t,k*t+d)
plt.show()
```

```python
#绘制数据草图
t=np.array([3,6,9,12,15,18,21,24])
y=np.array([57.6,41.9,31,22.7,16.6,12.2,8.9,6.5])
plt.plot(t,y)
plt.show()
```

```python
plt.grid()
plt.scatter(t,np.log10(y),c='r',s=40)
plt.show()
```

```python
a,b=sy.symbols('a b',real=True)
M=np.sum((np.log10(y)-(a*t+b))**2)
solves=sy.solve([M.diff(a),M.diff(b)],[a,b])
solves
```

```python
k=10**(float(solves[b]))
m=np.log(10)*(float(solves[a]))
m,k
```

```python
plt.grid()
plt.scatter(t,y,c='r',s=40)
x=np.linspace(2,25,100)
plt.plot(x,k*np.power(np.e,m*x))
plt.show()
```
