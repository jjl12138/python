```python
x,y=symbols('x y')
init_printing()
f=y**5+2*y-x-3*x**7
idiff(f,y,x)
```

```python
Y=solve(f.subs(x,0),y)
Y[0]
```

```python
k=idiff(f,y,x).subs([(x,0),(y,Y[0])])
k
```

```python
x,y=symbols('x y')
init_printing()
k=idiff(x**2/16+y**2/9-1,y,x).subs([(x,2),(y,3*sqrt(3)/2)])  #切线在该点处的切线斜率
#化简方程
simplify(y-3*sqrt(3)/2-(k*(x-2)))
```

```python
t,a,b=symbols('t a b')
x=a*cos(t)
y=b*sin(t)
init_printing()
diff_1_x=diff(y,t)/diff(x,t)
diff_1_x
```

```python
#二阶导数
diff_2_x=diff_1_x.diff(t)/x.diff(t)
simplify(diff_2_x)
```
