```python
x,y=symbols('x y')
init_printing()
f=y**5+2*y-x-3*x**7
idiff(f,y,x)
```

```python
Y=solve(f.subs(x,0),y)
Y[0]
```

```python
k=idiff(f,y,x).subs([(x,0),(y,Y[0])])
k
```

```python
t,a,b=symbols('t a b')
x=a*cos(t)
y=b*sin(t)
init_printing()
diff_1_x=diff(y,t)/diff(x,t)
diff_1_x
```
