```python
import matplotlib.pyplot as plt
X=[20,25,30,35,40,50,60,65,70,75,80,90]
y=[1.81,1.7,1.65,1.55,1.48,1.4,1.3,1.26,1.24,1.21,1.2,1.18]
plt.scatter(X,y)
plt.show()
```

```python
import numpy as np 
from sklearn import linear_model
squareX=[x**2 for x in X]
data=np.array([X,squareX]).T 
reg=linear_model.LinearRegression()
reg.fit(data,y)
a,b=reg.intercept_,reg.coef_
a,b
```

```python
plt.scatter(X,y)
X=np.array(X)
plt.plot(X,a+b[0]*X+b[1]*X**2,'r')
plt.show()
```
