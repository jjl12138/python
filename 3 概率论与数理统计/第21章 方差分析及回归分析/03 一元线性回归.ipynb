```python
import matplotlib.pyplot as plt
X=[100,110,120,130,140,150,160,170,180,190]
y=[45,51,54,61,66,70,74,78,85,89]
plt.scatter(X,y,color='r')
plt.show()
```

```python
import numpy as np 
from sklearn import linear_model
#将数据转换为一列
dataX=np.array(X).reshape(-1,1)
#建立线性回归模型
reg=linear_model.LinearRegression()
#使用线性回归模型拟合数据
reg.fit(dataX,y)
#输出截距与回归系数
a,b=reg.intercept_,reg.coef_
a,b
```

```python
plt.scatter(X,y,color='r')
plt.plot(X,a+b*np.array(X))
plt.show()
```

```python
X=list(range(1,11))
y=[2651,1943,1494,1087,765,538,484,290,226,204]
plt.scatter(X,y)
plt.show()
```

```python
y_prime=np.log(y)
plt.scatter(X,y_prime,color='r')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(np.array(X).reshape(-1,1),y_prime)
a,b=reg.intercept_,reg.coef_
a,b
```

```python
plt.scatter(X,y)
#y=np.e**(y_prime)=np.e**(a+b*X)实现数据的逆变换
plt.plot(X,np.e**(a+b*X),'r')
plt.show()
```
